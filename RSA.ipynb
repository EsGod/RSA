```python
keystart = time.time()

print("Gerando as chaves")
p=PrimeRNG(1024,2048)
print("p definido")
q=PrimeRNG(1024,2048)
while(p==q):
 q = PrimeRNG(1024,2048)
print("q definido")
n= getN(p,q)
print("n definido")
f= phi(p,q)
print("λ(n) definido")
e= getE(f)
print("e definido")
d= getD(e,f)
print("d definido")
privateKey = makePrivateKey(p,q,n,d)
print("chave privada criada")
publicKey = makePublicKey(e,n)
print("chave publica criada")
keyend = time.time()
print("Tempo gasto para geração das chaves : ",keyend - keystart)
```
